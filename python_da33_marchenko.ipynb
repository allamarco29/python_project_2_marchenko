Завдання 3. Статистичний аналіз досвіду респондентів



```python
df_work_exp = df_res['WorkExp'].dropna()

avg = df_work_exp.mean().round(2)
med = df_work_exp.median()
mod = df_work_exp.mode()[0]

print ('Міри центральної тенденції:' )
series = pd.Series([avg, med, mod], index=['Mean', 'Median', 'Mode'])
print(series)

#Співвідношення результатів:
#якщо Середнє (Mean) > Медіани (Median), то це означає, що серед даних є невелика група людей із високими показниками
#тобто, є певна кількість респондентів з дуже великим досвідом роботи, вони "тягнуть" середнє значення вгору
#надійнішим показником для аналітики в даному випадку буде значення Середнього (Median)
```

Завдання 5. Визначення популярності Python

```python
df_res['worked_with_python'] = df_res['LanguageHaveWorkedWith'].str.contains(r'\bPython\b', case=False, na=False)
print(df_res['worked_with_python'].value_counts())

#знаходимо кількість python-розробників за кількістю булевих значень True
```

```python
total_resp = len(df_res)
python_resp = df_res['worked_with_python'].sum()

perc_python = python_resp / total_resp * 100
print (f'Відсоток респондентів, які програмують на Python: {perc_python.round(2)} %')

#на Python програмують лише 37.43% опитаних респондентів або 18410 осіб
```

*Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників



```python
salary_higher75 = df_res['ConvertedCompYearly'].quantile(0.75)
salary_higher75

#визнаємо 75 перцентиль за компенсацією, результат - 120596,0 
```

```python
high_remote_workers = df_res[(df_res['ConvertedCompYearly'] >= salary_higher75) & (df_res['RemoteWork'] == 'Remote')]
high_remote_workers

#виокремлюємо респондентів, котрі працюють віддалено та мають зарплату більше ніж в інших 75% опитаних
```

```python
df_industry = high_remote_workers['Industry'].value_counts()
df_industry_remote = df_industry.reset_index()
df_industry_remote.columns = ['Industry', 'Count of remote workers']
df_industry_remote

#наводимо таблицю, згруповану по індустріях та кількістю респондентів (заплата >75% ніж в інших), котрі працюють віддалено
```
